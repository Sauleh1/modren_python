* OOP's:
    * OOPS, or More commonly called in programming terms as "Object Oriented Programming", Allows us to sort our data and organize it.
    * OOP's come with 2 types of methods, namely Magic Methods and Dunder Methods:
        * Magic Methods are methods with 2 "_" characters befpre and after them.
        * Dunder Methods are methods which do not have 2 "_" characters before and after them.
        * An example of a Magic method in use:
            ```
            class Student:
                Name = "Sauleh"
                Age = 12
                Education = "Sadequian Grammer School"
                def __init__(self, a: str, b: int, c: str) -> None:         <-- "__init__" is a magic method. It has 2 double Underscores
                    self.Name = a
                    self.Age = b
                    self.Education = c
            ```
        * An example of a Dunder method in use:
            ```
            class Student:
                Name = "Sauleh"
                Age = 12
                Education = "Sadequian Grammer School"
                def ExtractDetails(self) -> None:           <-- "ExtractDetails" is a dunder Method. It has no Underscores
                    a = input(f"Which details to extract from Students Details?\n1. {self.Name}\n2. {self.Age}\n3. {self.Education}")
                    if a == "1":
                        print(f"Name of student is {self.Name}")
                    elif a == "2":
                        print(f"Age of student is {self.Age}")
                    elif a == "3":
                        print(f"Education of student is {self.Education}")
            ```
    * OOP's come with 4 Types, namely:
        * Abstraction:
            * Abstraction in Python is achieved through abstract classes and interfaces. It's a core concept of Object-Oriented Programming   (OOP) that hides the complexity of a system, allowing users to implement methods without knowing the underlying logic. This      encapsulation of implementation details simplifies programming.
            * An example of abstraction code is:
                ```
                class Car:
                    def __init__(self, make, model, year):
                        self.make = make
                        self.model = model
                        self.year = year

                    def drive(self):
                        print(f"The {self.make} {self.model} {self.year} is driving.")

                tesla = Car("Tesla", "Model 3", 2023)

                tesla.drive()
                ```
        * Encapsulation:
            * Encapsulation is a fundamental concept in object-oriented programming (OOP) that describes the idea of wrapping data and the methods that work on data within one unit. This puts restrictions on accessing variables and methods directly and can prevent the accidental modification of data. To prevent accidental change, an object’s variable can only be changed by an object’s method. Those types of variables are known as private variables.
            * An example of Encapsulation code in python is:
            ```
            class Person:
                def __init__(self, name, age):
                    self.__name = name
                    self.__age = age

                def get_name(self):
                    return self.__name

                def get_age(self):
                    return self.__age

            person = Person("John Doe", 30)

            name = person.get_name()

            age = person.get_age()

            print(f"The person's name is {name} and their age is {age}")
            ```
        * Inheritance:
            * In Python, inheritance is a fundamental concept of Object-Oriented Programming (OOP) that allows a class to inherit all the methods and properties from another class¹. The class being inherited from is called the base or parent class, while the class that inherits is known as the derived or child class.
            * An example of Inheritance code in python is:
            ```
            class Animal:
                def eat(self):
                    print("The animal is eating.")

                def move(self):
                    print("The animal is moving.")

            class Dog(Animal):
                def bark(self):
                    print("The dog is barking.")

            dog = Dog()

            dog.eat()

            dog.move()

            dog.bark()
            ```
        * Polymorphisem:
            * Polymorphism in Python is a principle of object-oriented programming (OOP) that allows methods to behave differently based on the object that is calling them. It enables a single interface to represent different types in different scenarios.
            * An example of Inheritance code in python is:
            ```
            class Animal:
                def speak(self):
                    pass

            class Dog(Animal):
                def speak(self):
                    print("Woof!")

            class Cat(Animal):
                def speak(self):
                    print("Meow!")

            animals = [Dog(), Cat()]

            for animal in animals:
                animal.speak()
            ```
