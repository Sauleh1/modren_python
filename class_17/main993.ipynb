````python
import pandas as pd
import pandera
import numpy as np
from nptyping import NDArray, Shape, Int64

total = 10000

s1: NDArray[Shape[str(total) + "5"], Int64] = np.random.randint(80, 100, (total, 5))
ss1: pd.DataFrame = pd.DataFrame(s1, columns = ["s1","s2","s3","s4","s5"])

s2: NDArray[Shape[str(total) + "5"], Int64] = np.random.randint(80, 100, (total, 5))
ss2: pd.DataFrame = pd.DataFrame(s2, columns = ["s1","s2","s3","s4","s5"])

s3: NDArray[Shape[str(total) + "5"], Int64] = np.random.randint(80, 100, (total, 5))
ss3: pd.DataFrame = pd.DataFrame(s3, columns = ["s1","s2","s3","s4","s5"])

s4: NDArray[Shape[str(total) + "5"], Int64] = np.random.randint(80, 100, (total, 5))
ss4: pd.DataFrame = pd.DataFrame(s4, columns = ["s1","s2","s3","s4","s5"])

s5: NDArray[Shape[str(total) + "5"], Int64] = np.random.randint(80, 100, (total, 5))
ss5: pd.DataFrame = pd.DataFrame(s5, columns = ["s1","s2","s3","s4","s5"])

# print(len(ss1))
# print(len(ss2))
# print(len(ss3))
# print(len(ss4))
# print(len(ss5))

# ss1

df: pd.DataFrame = pd.concat([ss1])
print("================================================================")
display(df.info())
print("================================================================")
display(df.describe())
print("================================================================")
display(df.head())
````
